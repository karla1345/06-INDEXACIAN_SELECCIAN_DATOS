- ¿Cuál es el número máximo de pasajeros registrados? (Aplicar método max()). 

```python
ventas = pd.read_csv('Data/ventas.csv', encoding='latin-1')
```

- Calcula la media y la desviación estándar de la columna num_passengers. (Aplicar método mean() y std())

```python
#media_ventas_2022 = ventas_df['Ventas_2022'].mean()
Prom=ventas['num_passengers'].mean() # Calcula promedio 
desvstd=ventas['num_passengers'].std() # Calcula estandar
print(f"El promedio de numero de pasajeros es {Prom} \nLa desviación estandar es {desvstd}")
```

- ¿Cuál es el número máximo de pasajeros registrados? (Aplicar método max()).

```python
# max_ventas_2022 = ventas_df['Ventas_2022'].max()
print(f"Máximo de pasajeros en 2022: {ventas['num_passengers'].max()}")
```

# 6. Visualización de datos:

- Crea un gráfico de barras que muestre la cantidad de pasajeros por día de la semana `(flight_day)`.

```python
a=ventas['flight_day'].value_counts()
plt.figure(figsize=(10,6))  #tamaño de gráfico
plt.bar(a.index,a.values,color='red')

#añadir título y  etiquetas a los ejes.
plt.title('cantidad de pasajeros por dia')
plt.xlabel('dia')#poner etiqueta a eje x
plt.ylabel('Número de pasajeros')#poner etiqueta a eje y
plt.xticks(rotation=90)
plt.show()#mostrar Gráfico
```
